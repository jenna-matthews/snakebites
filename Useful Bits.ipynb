```
%sql
--to_date similar to trunc() in plsql; gets date in truncated (date only, no time) format - YYYY-MM-DD
select to_date(event_timestamp) as event_date from [tablename];
--last_day gets the last day of the month in truncated (date only, no time) format - YYYY-MM-DD
select last_day(the_date) as end_of_month from [tablename];
--max/min functions are similar to plsql
select max(the_date) as latest_date from [tablename];
```
